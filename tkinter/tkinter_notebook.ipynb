# Example-2

```python
import tkinter as tk

form = tk.Tk() # Tk class in dan bir GUI objesi olusturur 
form.title('Tkinter Examples') 

form.geometry('500x250+500+150') # GUI nin default x,y boyutlari ve desktopta acilacagi konum koordinatlari belirler
form.state('normal') # GUI, desktopta acilir

def function1():
    label= tk.Label(form, text= 'click_1', fg='red', bg='black') # fg:frontground, bg:background
    label.pack(side=tk.BOTTOM)

def function2():
    label= tk.Label(form, text= 'click_2', fg='blue', bg='black') # fg:frontground, bg:background
    label.pack(side=tk.TOP)

button1 = tk.Button(form, text= 'click_1', fg='black', bg='red', font= 'ariel 12 bold',command = function1)
button1.pack(side=tk.LEFT)

button2 = tk.Button(form, text= 'click_2', fg='black', bg='blue', font= 'ariel 12 bold',command = function2)
button2.pack(side=tk.RIGHT)

button3 = tk.Button(form, text= 'click_2', fg='black', bg='blue', font= 'ariel 12 bold',command = form.destroy)
button3.pack()

form.mainloop() # GUI yi acik tutar.
```

# Example-3
* 1-50 arasinda rastgele 5 sayi uretsin,
* show butonu ile ekrana yazdirsin,
* transparent butonu ile ekran saydamlastirsin
* default butonu ile ayarlar default hale gelsin
* full screen butonu ie tam ekran yap,
* min screen butonu ile kucuk ekran yap

```python
from random import sample
import tkinter as tk

form = tk.Tk()
form.title('Example-3')
form.geometry('420x250+500+250')
form.state('normal')

def show_numbers():
    label.config(text= str(sorted(sample(range(1,51),5))))
    
def transparent_screen():
    form.wm_attributes('-alpha',0.6)    
    
def default_screen():
    form.wm_attributes('-alpha',1)  
    
def full_screen():
    form.state('zoomed')
    
def min_screen():
    form.state('normal')
    form.geometry('420x250+500+250')
    
label= tk.Label(form, fg='blue', bg='grey',font= 'ariel 17 bold')
label.pack()  

show_button = tk.Button(form, text= 'show', fg='black', bg='red', font= 'ariel 12 bold',command = show_numbers)
show_button.pack(side=tk.LEFT)

transparent_button = tk.Button(form, text= 'transparent', fg='white', bg='limegreen', font= 'ariel 12 bold',command = transparent_screen)
transparent_button.pack(side=tk.LEFT)

default_button = tk.Button(form, text= 'default', fg='black', bg='darkgreen', font= 'ariel 12 bold',command = default_screen)
default_button.pack(side=tk.LEFT)

min_button = tk.Button(form, text= 'min screen', fg='grey', bg='maroon', font= 'ariel 12 bold',command = min_screen)
min_button.pack(side=tk.LEFT)

full_button = tk.Button(form, text= 'full screen', fg='grey', bg='navy', font= 'ariel 12 bold',command = full_screen)
full_button.pack(side=tk.LEFT)

form.mainloop()
```

# Example-6
* User Name ve Password u alip ekrana yazdiran,
* remove butonu ile formlari temizleyen app yapiniz.

```python
import tkinter as tk

form = tk.Tk()
form.title('Tkinter Examples') 

form.geometry('500x400+500+150')
form.state('normal')

user_label= tk.Label(form, text='User Name', fg='black',font= 'ariel 17 bold') 
user_label.pack()  

user_name = tk.Entry(form, fg='black',font= 'ariel 17 normal') 
user_name.pack()

password_label= tk.Label(form, text='Password', fg='black',font= 'ariel 17 bold') 
password_label.pack()

password = tk.Entry(form, show='*',fg='black',font= 'ariel 17 normal') 
password.pack()

def get_data():
    label['text'] = f"Username: {user_name.get()}\nPassword: {password.get()}" 
    
def remove():
    user_name.delete(0,'end')
    password.delete(0,'end')
    
label= tk.Label(form, text='Lorem ipsum dolor sit amet', fg='blue',font= 'ariel 17 bold') 
label.pack() 

button_1 = tk.Button(form, text= 'GET DATA', fg='black', bg='green', font= 'ariel 18 bold',command = get_data)
button_1.pack(anchor='s',
            fill=tk.X,
            padx=10,pady=20,
            ipady=10)

button_2 = tk.Button(form, text= 'REMOVE ENTRY', fg='black', bg='red', font= 'ariel 18 bold',command = remove)
button_2.pack(anchor='s',
            fill=tk.X,
            padx=10,pady=20,
            ipady=10) # butonun x ve y boyutlarini belirler.

form.mainloop()
```

# Check Button

```python
import tkinter as tk

form = tk.Tk()
form.title('Check Button') 

form.geometry('500x250+500+150') # GUI nin default x,y boyutlari ve desktopta acilacagi konum koordinatlari belirler
form.state('normal') # GUI, desktopta acilir

x=tk.IntVar()
x.set(0)

def check():
    if x.get()==1:
        label= tk.Label(form, text= 'validate', fg='green', bg='black') # fg:frontground, bg:background
        label.pack()
    else:
        label= tk.Label(form, text= 'unvalidate', fg='red', bg='black') # fg:frontground, bg:background
        label.pack()

check_button = tk.Checkbutton(form, text= 'check box', fg='black', bg='blue', font= 'ariel 12 bold',variable=x, command = check)
check_button.pack(side=tk.LEFT)

form.mainloop() # GUI yi acik tutar.
```

# Radio Button

```python
import tkinter as tk

form = tk.Tk()
form.title('Radio Button') 

form.geometry('500x250+500+150') # GUI nin default x,y boyutlari ve desktopta acilacagi konum koordinatlari belirler
form.state('normal') # GUI, desktopta acilir

x=tk.StringVar()

def check():
    if x.get()=='1':
        label= tk.Label(form, text= 'radio_1', fg='red', bg='black')
        label.pack()
    elif x.get()=='2':
        label= tk.Label(form, text= 'radio_2', fg='green', bg='black')
        label.pack()
    else:
        label= tk.Label(form, text= 'radio_1 and radio_2', fg='blue', bg='black')
        label.pack()
       

button = tk.Button(form, text= 'click', fg='black', bg='blue', font= 'ariel 12 bold',activebackground='green', command = check)
button.pack(side=tk.LEFT)        

radio_1 = tk.Radiobutton(form, text= 'radio_1',activebackground='red',value=1, variable=x)
radio_1.pack()

radio_2 = tk.Radiobutton(form, text= 'radio_2',activebackground='green',value=2, variable=x)
radio_2.pack()

form.mainloop()
```

# tkinter with OOP

```python
from tkinter import Frame,Tk,Label,Text,INSERT, END

class Frame1(Frame):
    def __init__(self, parent):
        Frame.__init__(self, parent, bg="red")
        self.parent = parent
#         self.parent.geometry('300x400+500+150')
#         self.parent.state('normal')
        self.widgets()

    def widgets(self):
        self.text = Text(self) # Text classindan obje olusturduk.
        self.text.insert(INSERT, "Hello World\n")
        self.text.insert(END, "This is the first frame")
        self.text.grid(row=0, column=0, 
#                        padx=10, pady=10, # text objesinin cerceve margin ini belirledik.
                      )

class MainW(Tk):

    def __init__(self, parent):
        Tk.__init__(self, parent)
        self.parent = parent
        self.mainWidgets()

    def mainWidgets(self):

        self.label1 = Label(self, text="Main window label", bg="green")
        self.label1.grid(row=0, column=0, sticky = N) 

        self.label2 = Label(self, text="Main window label", bg="yellow")
        self.label2.grid(row=1, column=0, sticky = N)

        self.window = Frame1(self) # MainW classi icinde Frame1 classindan obje olusturduk. Bir class icinde baska classi kullandik.
        self.window.grid(row=0, column=1, rowspan=2)

# if __name__=="__main__":
#     app = MainW(None)
#     app.mainloop()

app = MainW(None)
app.mainloop()
```

# forget() & pack()

```python
from tkinter import * 
from tkinter.ttk import *
  
# toplevel window 
root = Tk() 
  
# Button widgets
b1 = Button(root, text = "Btn 1") 
b1.pack(fill = BOTH, expand = True) 
  
# See, in command forget() method is passed 
b2 = Button(root, text = "Btn 2", command = lambda : b1.forget()) 
b2.pack(fill = BOTH, expand = True) 
  
# In command retrieve() method is passed 
b3 = Button(root, text = "Btn 3", command = lambda : b1.pack(fill = BOTH, expand = True) ) 
b3.pack(fill = BOTH, expand = True) 
  
# infinite loop, interrupted by keyboard or mouse 
mainloop() 
```

```python
import tkinter as tk


class TestGrid(tk.Frame):
    def __init__(self, master, switch):
        tk.Frame.__init__(self, master, bg='white')
        colors = ['cyan', 'magenta', 'green', 'gold', 'lavender', 'purple']
        self.switch = switch

        self.widgets = []
        for i in range(6):
            self.widgets.append(tk.Label(self, text=i, bg=colors[i]))

    def draw(self):
        for w in self.widgets:
            w.grid_forget()

        positions = [(0, 0), (1, 0), (2, 0), (0, 1), (1, 1), (2, 1)]
        for i in range(6):
            c, r = positions[i][0], positions[i][1]
            if self.switch:
                c, r = r, c
            self.grid_columnconfigure(c, weight=1, uniform="aaa")
            self.grid_rowconfigure(r, weight=1, uniform="bbb")
            self.widgets[i].grid(column=c, row=r, sticky=tk.NSEW)


class MyApp(tk.Tk):
    def __init__(self):
        tk.Tk.__init__(self)
        self.geometry('400x400+20+20')

        btn = tk.Button(self, text="click me")
        btn.pack(side=tk.TOP, fill=tk.X)
        btn.bind("<Button-1>", self.transpose)

        self.switch = False
        self.frm = None
        self.create_frm()

    def create_frm(self):
        self.frm = TestGrid(self, self.switch)
        self.frm.draw()
        self.frm.pack(fill=tk.BOTH, expand=1, padx=5, pady=5)

    def transpose(self, *args):
        self.frm.destroy()
        self.switch = not self.switch
        self.create_frm()


def main():
    root = MyApp()
    root.mainloop()


if __name__ == '__main__':
    main()
```

# Scrollbar

```python
from tkinter import *
from tkinter.ttk import *

# http://tkinter.unpythonic.net/wiki/VerticalScrolledFrame

class VerticalScrolledFrame(Frame):
    """A pure Tkinter scrollable frame that actually works!
    * Use the 'interior' attribute to place widgets inside the scrollable frame
    * Construct and pack/place/grid normally
    * This frame only allows vertical scrolling

    """
    def __init__(self, parent, *args, **kw):
        Frame.__init__(self, parent, *args, **kw)            

        # create a canvas object and a vertical scrollbar for scrolling it
        vscrollbar = Scrollbar(self, orient=VERTICAL)
        vscrollbar.pack(fill=Y, side=RIGHT, expand=FALSE)
        canvas = Canvas(self, bd=0, highlightthickness=0,
                        yscrollcommand=vscrollbar.set)
        canvas.pack(side=LEFT, fill=BOTH, expand=TRUE)
        vscrollbar.config(command=canvas.yview)

        # reset the view
        canvas.xview_moveto(0)
        canvas.yview_moveto(0)

        # create a frame inside the canvas which will be scrolled with it
        self.interior = interior = Frame(canvas)
        interior_id = canvas.create_window(0, 0, window=interior,
                                           anchor=NW)

        # track changes to the canvas and frame width and sync them,
        # also updating the scrollbar
        def _configure_interior(event):
            # update the scrollbars to match the size of the inner frame
            size = (interior.winfo_reqwidth(), interior.winfo_reqheight())
            canvas.config(scrollregion="0 0 %s %s" % size)
            if interior.winfo_reqwidth() != canvas.winfo_width():
                # update the canvas's width to fit the inner frame
                canvas.config(width=interior.winfo_reqwidth())
        interior.bind('<Configure>', _configure_interior)

        def _configure_canvas(event):
            if interior.winfo_reqwidth() != canvas.winfo_width():
                # update the inner frame's width to fill the canvas
                canvas.itemconfigure(interior_id, width=canvas.winfo_width())
        canvas.bind('<Configure>', _configure_canvas)


if __name__ == "__main__":

    class SampleApp(Tk):
        def __init__(self, *args, **kwargs):
            root = Tk.__init__(self, *args, **kwargs)


            self.frame = VerticalScrolledFrame(root)
            self.frame.pack()
            self.label = Label(text="Shrink the window to activate the scrollbar.")
            self.label.pack()
            buttons = []
            for i in range(10):
                buttons.append(Button(self.frame.interior, text="Button " + str(i)))
                buttons[-1].pack()

    app = SampleApp()
    app.mainloop()
```

# after()

```python
from tkinter import Tk, mainloop, TOP 
from tkinter.ttk import Button 
  
# time function used to calculate time 
from time import time 
  
# creating tkinter window 
root = Tk() 
root.geometry("500x400+500+250")

button = Button(root, text = 'Geeks') 
button.pack(side = TOP, pady = 5) 
  
print('Running...') 
# Calculating starting time 
start = time() 
  
# in after method 5000 miliseconds 
# is passed i.e after 5 seconds 
# main window i.e root window will 
# get destroyed 
root.after(5000, root.destroy) 
  
mainloop() 
  
# calculating end time 
end = time() 
print('Destroyed after % d seconds' % (end-start)) 
```
